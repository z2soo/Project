```
# 데이터 프레임 칼럼안에 눌 데이터가 있는 지 확인: 없음

def check_if_null(train):
    if train.isnull().values.any() or "No Data" in train.values or "Bad" in train.values or "CLOSE" in train.values or "Equip Fail" in train.values or "Normal" in train.values or "OFF" in train.values or "ON" in train.values or "OPEN" in train.values or "System.Char[]" in train.values :
        print("DataFrame has certain Values!")
        return True
```

### column 크기 줄이기

```
print(train.shape)
print(X_train.shape)
```

```
# 동일한 값을 가지는 column 삭제

def del_cols_same_val(X_train):
    n_unique = X_train.apply(pd.Series.nunique)
    cols_to_drop = n_unique[n_unique == 1].index
    X_train.drop(cols_to_drop, axis = 1, inplace = True)
    return X_train

del_cols_same_val(train)
print(train.shape)
```

```
# 동일 값을 가지는 col 삭제 데이터 저장

train.to_csv('train_del_col')
```

```
def del_row_same_val(X_train):
    n_unique = X_train.apply(pd.Series.nunique)
    cols_to_drop = n_unique[n_unique == 1].index
    X_train.drop(cols_to_drop, axis = 1, inplace = True)
    return X_train
del_row_same_val(X_train)

print(train.shape)
print(X_train.shape)
```
