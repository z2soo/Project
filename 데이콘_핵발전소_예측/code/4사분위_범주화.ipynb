```
# 모듈 삽입

import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt 
from tensorflow.examples.tutorials.mnist import input_data
```

```
# 드라이브에 저장된 데이터 불러오기

data_train_total = pd.read_csv("/content/drive/My Drive/Colab Notebooks/project_nuclear/project_processed_data/output/B_train_except30.csv")
df = data_train_total
```

```
df.shape[1]
```

```
for i in range(1,df.shape[1]):
    q1,q2,q3 = np.percentile(df.iloc[:,i],[25,50,75])
    df.loc[df.iloc[:,i] <= q1, df.columns[i]] = 0
    df.loc[(df.iloc[:,i] > q1) & (df.iloc[:,i] <= q2), df.columns[i]] = 1
    df.loc[(df.iloc[:,i]> q2) & (df.iloc[:,i] <= q3), df.columns[i]] = 2
    df.loc[df.iloc[:,i] > q3, df.columns[i]] = 3
print(df)
```
