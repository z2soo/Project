## attribute_labels 해석 : image가 가진 attribute_name 알아내기, column 생성

```python
df = total_data[['image_name', 'attribute_labels']]
df
```

```python
# image가 가지는 attribute_name를 알아내서 새로운 col 값으로 만들기


# attribute_labels 값을 리스트로 생성
# attrubute_name 값을 리스트로 생성
# 두 리스트를 비교해서 '1' 값을 가지는 경우는 해당 attribute를 가지는 경우 : attribute_name 출력하도록 함


# 1. attribute_name 값을 저장할 column 생성
df['attribute_name'] = 'NA'


# 2. attribute_name 값 알려주는 함수 생성
def my_name(x):
    my_list = list(df['attribute_labels'][x].split(' '))     # 해석하려는 attribute_labels 값
    name_list = list(at_cloth['attribute_name'])             # 비교하려는 attribute_name 값
    result = []
    
    for _ in range(len(my_list)):
        if my_list[_] != '-1':
            pass
        elif my_list[_] == '1':
            n = name_list[_]
            result.append(n)       
    df['attribute_name'][x] = result     
    
    
```

```python
# 3. 전체 attribute_labels에 적용
for i in range(df.shape[0]):
    my_name(i)
```
