## 3. attribute 파일 2개: 불필요 col 제거, 공백 제거, 파일로 저장

```python
# 공백 제거 함수 생성

def remove(x):
    cols = x.columns
    print(cols)
    for col in cols:
        for _ in range(x.shape[0]):
            if type(x[col][_]) == str:
                x[col][_] = x[col][_].strip()
```

```python
# 1. at_cloth 


# 불필요 col 제거
at_cloth = at_cloth.drop('attribute_name  attribute_type', axis=1)

# 공백제거
remove(at_cloth)

# 확인
at_cloth.columns
```

```python
# 파일로 저장

at_cloth.to_csv('attr_cloth_final.csv', header=True, index=False)
```

```python
# 2. at_img


# 불필요 col 제거
at_img = at_img.drop('image_name  attribute_labels', axis=1)

# 공백제거
remove(at_img)

# 확인
at_img.columns
```
