```python
# 원본 파일 열어보면 맨 위에 나온 숫자 = 행의 갯수
# 직접 지워서 사용함

# 1. list_attr_cloth 파일 

# attribution name / attribution type
# 1000개 attribution name과 1~5 숫자로 표현되는 attribution type

df_1 = pd.read_csv('E:\Project\list_attr_cloth.txt')
display(df_1)

df_1.columns #['attribute_name  attribute_type']
df_1['attribute_name'] = df_1['attribute_name  attribute_type'].str[:-1]
df_1['attribute_type'] = df_1['attribute_name  attribute_type'].str[-1]
display(df_1)
print(df_1.shape)
df_1.to_csv('E:\Project\list_attr_cloth.csv', header=True, index=False)
```
