## 3. 이미지 데이터 폴더명/이미지명 구분

```python
# total_data_final 데이터프레임의 image_name
# img/폴더명/이미지명 : 이렇게 저장되어 있음

# 이미지 불러오기 편하게 폴더명, 이미지명 column 생성하기
```

```python
# 데이터 파일 불러와서 폴더명, 이미지 이름 구분해서 column 생성

df['folder_name'] = 'NA'
df['image_num'] = 'NA'
```

```python
# 폴더명, 이미지 이름 구분한 값 위에서 생성한 column 값으로 넣어주기

for _ in range(df.shape[0]):
    s = df['image_name'][_].split('/')
    folder = s[1]
    df['folder_name'][_] =folder
    image_num = s[2]
    df['image_num'][_] = image_num
    
    if _ % 1000 == 0:
        print(_)
```

```python
# column 순서 보기 편하게 변경

df.columns
df = df[['image_name','folder_name','image_num','category_label', 'category_name','category_type','attribute_labels','texture', 'shape', 'style']]
```
