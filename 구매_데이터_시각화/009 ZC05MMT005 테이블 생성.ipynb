### 한국어로 번역

```python
# 아래 코드는 계속 다시 번역하기 때문에 시간이 너무 걸림
# 따라서 20개 번역한 값을 저장해서 이를 넣어주도록 코드 수정함 

# for _ in range(len(df05)):
#     source = df05['COMT'][_]
#     translator = Translator()
#     target = translator.translate(source, src='en', dest='ko')
#     df05['COMTK'][_] = target.text
#     if _ % 1 == 0:                #잘 돌아가는지 중간 확인용
#         print(_)
#         print(target.text)


k = 0
for i in range(len(df05)):
    if i < 90:                     #comment 한 싸이클이 90개이기 때문
        source = df05['COMT'][i]
        translator = Translator()
        target = translator.translate(source, src='en', dest='ko')
        df05['COMTK'][i] = target.text
    elif i >= 90:
        df05['COMTK'][i] = df05['COMTK'][k]
        k += 1
    
    if i % 20 == 0:                 #잘 돌아가는지 중간 확인용
        print(i)
        print(target.text)
```

```python
result05 = pd.DataFrame(df05)

# xlsx 파일로 저장
result05.to_excel(excel_writer='ZC05MMT005 한국어 번역.xlsx', header=True, index=False)
```
