### ZC05MMT04 테이블 생성

```python
df01 = pd.read_csv("ZC05MMT001.csv")
display(df01.head())

df02 = pd.read_csv("ZC05MMT002.csv")
display(df02.head())

df04 = pd.read_csv("ZC05MMT004 layout.csv")
display(df04.head())
```

```python
# 공급업체 이름 추가 

sourceList = df02['LIFNR']

if len(df04) != 3600:
    for _ in range(len(sourceList)):
        df04 = df04.append({'LIFNR' : sourceList[_]}, ignore_index=True)
else:
    print(f'이미 공급업체명이 추가됬습니다. 공급업체 리스트는 총 {len(df04)}개 입니다.')
    display(df04.head())
```

```python
# 공급업체 주소 정보 추가

for _ in range(len(df04)):
    var1 = df04['LIFNR'][_]
    i = 0
    while True:
        if df01['LIFNR'][i] == var1:
            df04['NAME1'][_] = df01['NAME1'][i]
            df04['LAND1'][_] = df01['LAND1'][i]
            df04['ORT01'][_] = df01['ORT01'][i]
            break
        else:
            i+=1
```
